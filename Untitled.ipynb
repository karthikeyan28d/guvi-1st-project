```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
```

```python
df=df[~((df<(q1-1.5*iqr))|(df>(q3+1.5*iqr))).any(axis=1)]
```

```python
# 3SsnPorch	NaN
# 5	BsmtFinSF2	NaN
# 7	BsmtHalfBath	NaN
# 9	EnclosedPorch	NaN
# 20	LowQualFinSF	NaN
# 23	MiscVal	NaN
# 28	PoolArea	NaN
# 29	ScreenPorch	
```

```python
df=df.drop(["3SsnPorch","BsmtFinSF2","BsmtHalfBath","EnclosedPorch","LowQualFinSF","MiscVal","PoolArea","ScreenPorch"],axis=1)
```

```python
new=df.select_dtypes(include=np.number).drop(["SalePrice"],axis=1)
```

```python
vif=pd.DataFrame()
```

```python
vif["feature"]=new.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif["score"]=[variance_inflation_factor(new.values,i) for i in range(new.shape[1])]
```

```python
vif.sort_values("score",ascending=False)
```

```python
cat=df.select_dtypes(include="object")
```

```python
df_cat=pd.get_dummies(cat,drop_first=True)
```

```python
df_cat.head()
```

```python
df_cat.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
num=df.select_dtypes(include=np.number)
```

```python
num.head()
```

```python
df_num=pd.DataFrame(ss.fit_transform(num),columns=num.columns,index=num.index)
```

```python
df_num.head()
```

```python
x=pd.concat([df_num,df_cat],axis=1).drop(["SalePrice"],axis=1)
```

```python
x.head()
```

```python
x.shape
```

```python
y=df["SalePrice"]
```

```python
y.shape
```

```python
import statsmodels.api as sm
from sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=110,test_size=0.3)
```

```python
model=sm.OLS(y_train,x_train).fit()
```

```python
model.summary()
```

```python
model.pvalues[model.pvalues<0.05]
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(x_train.corr()[np.abs(x_train.corr()>0.5)],annot=True)
```

```python
vif.set_index("feature",inplace=True)
```

```python
sig_feat=vif[vif["score"]<4].index.to_list()
```

```python
sig_feat
```

```python
model=sm.OLS(y_train,x_train[sig_feat]).fit()
```

```python
model.summary()
```

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
plt.rcParams["figure.figsize"]=(15,8)
qqplot(model.resid,line="r")
plt.title("qqplot")
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_pred_train=model.predict(x_train[sig_feat])
y_pred_test=model.predict(x_test[sig_feat])
```

```python
mse_train=mean_squared_error(y_train,y_pred_train)
mse_test=mean_squared_error(y_test,y_pred_test)
```

```python
print(mse_train)
print(mse_test)
```

```python
print(np.sqrt(mse_train))
print(np.sqrt(mse_test))
```
